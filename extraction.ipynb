```python
from PyPDF2 import PdfReader
reader = PdfReader("./book.pdf")
```

```python
def cleaned(x:str):
    word = []
    meaning = []
    sentence = []
    datalist = x.split('\n \n')
    for pos,val in enumerate(datalist):
        if pos+1 != len(datalist):
            if '=' in val:
                ls = val.replace('\n',' ').split('=')
                word.append(ls[0])
                meaning.append(ls[1])
                sentence.append(datalist[pos+1].replace('\n',' '))
        
    return (word,sentence)
```

```python
def to_dict():
    pagenumbers = range(len(reader.pages))
    dictionary = {'Word':[],
                  'Meaning':[],
                  'Sentence':[]}
    for i in pagenumbers:
        page = reader.pages[i]
        x = page.extract_text()
        newlist  = cleaned(x)
        dictionary['Word'].extend(newlist[0])
        dictionary['Sentence'].extend(newlist[1])
    return dictionary
```

```python
x = to_dict()
```

```python
import pandas as pd
dataframe = pd.DataFrame.from_dict(x)
```
